```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("resultsv9.csv")
```

```python
df.head()
```

```python
# Filter columns that do not start with 'z' or '_'
filtered_columns = [col for col in df.columns if not (col.startswith('z') or col.startswith('_'))]

num_vars = len(filtered_columns)
dist = 50

# Create subplots
fig, axes = plt.subplots(num_vars, 1, figsize=(10, 6 * num_vars), sharex=True)

# Loop through each filtered column and plot its values with a smooth black line showing the 10-point moving average
for i, column in enumerate(filtered_columns):
    subset = df[column]
    axes[i].plot(subset, label=column)
    # Calculate and plot the moving average
    moving_avg = subset.rolling(window=dist).mean()
    axes[i].plot(moving_avg, color='black', label=f'{dist} Point Moving Average')
    axes[i].set_title(f'Column: {column}')
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel('Value')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
